## Complex Numbers: Roots of a Complex Number

This notebook demonstrates how to compute the **nth roots** of a complex number using Python and visualize them.

### Example:
Find the 4th roots of the complex number `z = 1 + j`.



**Step 1: Convert to Polar Form**

A complex number `z = a + bj` can be represented in polar form as:
- Magnitude: `|z| = sqrt(a² + b²)`
- Angle: `θ = arctan(b/a)`


**Step 2: Compute Roots Using De Moivre's Theorem**

$$
z_k = \rho^{1/n} \cdot e^{j\left(\frac{\theta + 2\pi k}{n}\right)}, \quad k = 0, 1, \dots, n-1
$$

**Step 3: Plot the Roots on the Complex Plane**

We use matplotlib to visualize the roots with arrows (like MATLAB's `compass` plot).


```python
import numpy as np
import matplotlib.pyplot as plt

# Input complex number
x1 = 1
y1 = 1
z1 = complex(x1, y1)
n = 4  # nth root

# Convert to polar
rho = abs(z1)
theta = np.angle(z1)

# Magnitude of each root
mag = rho ** (1/n)

# Compute roots using De Moivre's Theorem
roots = []
for k in range(n):
    angle_k = (theta + 2 * np.pi * k) / n
    zk = mag * np.exp(1j * angle_k)
    roots.append(zk)

# Display roots
print(f"Original Complex Number: {z1}")
print(f"{n}th Roots:")
for i, r in enumerate(roots):
    print(f"Root {i+1}: {r:.4f}")

# Plotting using matplotlib (compass-style)
plt.figure(figsize=(6,6))
ax = plt.subplot(111, polar=True)
for r in roots:
    ax.plot([0, np.angle(r)], [0, abs(r)], marker='o', label=f'{r:.2f}')
ax.set_title(f'{n}th Roots of {z1}')
ax.legend(loc='upper right')
plt.show()
```

### 1. **Addition of Complex Numbers**  
**Explanation**:  
Addition of two complex numbers is performed by adding their real parts and their imaginary parts separately. For example, when adding \(z_1 = 3 + 4j\) and \(z_2 = 1 - 2j\), you add the real parts (3 and 1) and the imaginary parts (4 and -2) individually.

**Mathematical Example**:
Let:
$$
z_1 = 3 + 4j, \quad z_2 = 1 - 2j
$$
Then:
$$
z_3 = z_1 + z_2 = (3 + 4j) + (1 - 2j) = 4 + 2j
$$
Thus, the sum of \(z_1\) and \(z_2\) is \(4 + 2j\).

---

### 2. **Subtraction of Complex Numbers**  
**Explanation**:  
Subtraction of complex numbers involves subtracting the real parts and the imaginary parts separately. This is similar to addition, except we subtract rather than add.

**Mathematical Example**:
Let:
$$
z_1 = 3 + 4j, \quad z_2 = 1 - 2j
$$
Then:
$$
z_4 = z_1 - z_2 = (3 + 4j) - (1 - 2j) = 2 + 6j
$$
Thus, the result of subtracting \(z_2\) from \(z_1\) is \(2 + 6j\).

---

### 3. **Multiplication of Complex Numbers**  
**Explanation**:  
To multiply two complex numbers, you use the distributive property (also known as FOIL for binomials) to multiply the real and imaginary parts. Additionally, remember that \(j^2 = -1\) to simplify the result.

**Mathematical Example**:
Let:
$$
z_1 = 3 + 4j, \quad z_2 = 1 - 2j
$$
Then:
$$
z_5 = z_1 \times z_2 = (3 + 4j) \times (1 - 2j)
$$
Expanding this:
$$
z_5 = (3)(1) + (3)(-2j) + (4j)(1) + (4j)(-2j)
$$
$$
z_5 = 3 - 6j + 4j - 8j^2
$$
Since \(j^2 = -1\), we get:
$$
z_5 = 3 - 6j + 4j + 8 = 11 - 2j
$$
Thus, the product of \(z_1\) and \(z_2\) is \(11 - 2j\).

---

### 4. **Division of Complex Numbers**  
**Explanation**:  
To divide one complex number by another, you multiply both the numerator and the denominator by the complex conjugate of the denominator. This eliminates the imaginary part in the denominator and simplifies the expression.

**Mathematical Example**:
Let:
$$
z_1 = 3 + 4j, \quad z_2 = 1 - 2j
$$
Then:
$$
z_6 = \frac{z_1}{z_2} = \frac{(3 + 4j)}{(1 - 2j)}
$$
We multiply both the numerator and denominator by the conjugate of \(z_2\), which is \(1 + 2j\):
$$
z_6 = \frac{(3 + 4j)(1 + 2j)}{(1 - 2j)(1 + 2j)} = \frac{(3 + 4j)(1 + 2j)}{1^2 + 2^2}
$$
Simplifying the denominator:
$$
z_6 = \frac{(3 + 4j)(1 + 2j)}{5}
$$
Expanding the numerator:
$$
z_6 = \frac{3 + 6j + 4j + 8j^2}{5} = \frac{3 + 10j - 8}{5} = \frac{-5 + 10j}{5}
$$
Thus:
$$
z_6 = -1 + 2j
$$
Thus, the division of \(z_1\) by \(z_2\) is \(-1 + 2j\).

---

### 5. **Complex Conjugate**  
**Explanation**:  
The conjugate of a complex number is obtained by changing the sign of the imaginary part. This operation is useful in division to eliminate the imaginary part in the denominator and simplify expressions.

**Mathematical Example**:
Let \(z_1 = 3 + 4j\). The conjugate is:
$$
z_1^* = 3 - 4j
$$
Thus, the conjugate of \(z_1\) is \(3 - 4j\).

---

### 6. **Opposite of a Complex Number**  
**Explanation**:  
The opposite of a complex number is simply the negation of both its real and imaginary parts. This operation reflects the complex number across the origin on the complex plane.

**Mathematical Example**:
Let \(z_1 = 3 + 4j\). The opposite is:
$$
-z_1 = -3 - 4j
$$
Thus, the opposite of \(z_1\) is \(-3 - 4j\).

---

### 7. **Polar Form of a Complex Number**  
**Explanation**:  
A complex number can be expressed in polar form, where its magnitude (or modulus) \(r\) is the distance from the origin to the point on the complex plane, and its angle \(\theta\) is the angle it makes with the positive real axis. The polar form is given by:
$$
z = r \cdot \exp(j\theta)
$$
Where:
- \(r = |z|\) is the magnitude,
- \(\theta = \arg(z)\) is the argument (or phase).

**Mathematical Example**:
For \(z_1 = 3 + 4j\), the magnitude is:
$$
r_1 = |z_1| = \sqrt{3^2 + 4^2} = 5
$$
And the angle is:
$$
\theta_1 = \arg(z_1) = \tan^{-1}\left(\frac{4}{3}\right) \approx 0.93 \, \text{radians}
$$
Thus, the polar form of \(z_1\) is:
$$
z_1 = 5 \cdot \exp(j \cdot 0.93)
$$

---

### 8. **Multiplication and Division in Polar Form**  
**Explanation**:  
When multiplying two complex numbers in polar form, you multiply their magnitudes and add their angles. When dividing, you divide their magnitudes and subtract their angles.

**Mathematical Example**:
For multiplication:
Let:
$$
z_1 = 5 \cdot \exp(j \cdot 0.93), \quad z_2 = 2 \cdot \exp(j \cdot 1.57)
$$
The product is:
$$
z_3 = 5 \times 2 \cdot \exp(j \cdot (0.93 + 1.57)) = 10 \cdot \exp(j \cdot 2.5)
$$
Thus, the result of multiplying \(z_1\) and \(z_2\) is \(10 \cdot \exp(j \cdot 2.5)\).

For division:
$$
z_3 = \frac{z_1}{z_2} = \frac{5 \cdot \exp(j \cdot 0.93)}{2 \cdot \exp(j \cdot 1.57)}
$$
The result is:
$$
z_3 = \frac{5}{2} \cdot \exp(j \cdot (0.93 - 1.57)) = 2.5 \cdot \exp(j \cdot -0.64)
$$
Thus, the result of dividing \(z_1\) by \(z_2\) is \(2.5 \cdot \exp(j \cdot -0.64)\).

---

### 9. **Exponentiation of Complex Numbers**  
**Explanation**:  
Exponentiation of complex numbers is typically done using polar form. If a complex number is expressed as \(z = r \cdot \exp(j\theta)\), then raising it to a power or taking a root involves modifying the magnitude and the angle.

**Mathematical Example**:
For \(z_7 = 0.707 + 0.707j\), we have:
$$
z_7^4 = (0.707 + 0.707j)^4 = 1 \cdot \exp(j \cdot 4\cdot 0.785) = \exp(j \cdot 3.14) = -1
$$
Similarly:
$$
z_7^{1/4} = \exp(j \cdot \frac{0.785}{4}) \approx 0.923 + 0.382j
$$
Thus, \(z_7^4 = -1\) and \(z_7^{1/4} \approx 0.923 + 0.382j\).


```python
import numpy as np
import matplotlib.pyplot as plt

# Function to plot a complex number
def plot_complex_numbers(numbers, colors, labels, title):
    plt.figure()
    for i, z in enumerate(numbers):
        plt.quiver(0, 0, np.real(z), np.imag(z), angles='xy', scale_units='xy', scale=1, color=colors[i])
        plt.text(np.real(z), np.imag(z), labels[i], fontsize=12, ha='right')
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.axhline(0, color='black',linewidth=1)
    plt.axvline(0, color='black',linewidth=1)
    plt.grid(True)
    plt.title(title)
    plt.show()

# 1. Complex numbers in Python
print('Exercises with complex numbers')
z1 = 3 + 4j
z2 = 1 - 2j

print(f'z1 = {z1}')
print(f'z2 = {z2}')

# 2. Addition
z3 = z1 + z2
print(f'z3 = z1 + z2 = {z3}')
plot_complex_numbers([z1, z2, z3], ['b', 'r', 'g'], ['z1', 'z2', 'z3=z1+z2'], 'Addition of Complex Numbers')

# 3. Subtraction
z4 = z1 - z2
print(f'z4 = z1 - z2 = {z4}')
plot_complex_numbers([z1, z2, z4], ['b', 'r', 'g'], ['z1', 'z2', 'z4=z1-z2'], 'Subtraction of Complex Numbers')

# 4. Multiplication
z5 = z1 * z2
print(f'z5 = z1 * z2 = {z5}')
plot_complex_numbers([z1, z2, z5], ['b', 'r', 'g'], ['z1', 'z2', 'z5=z1*z2'], 'Multiplication of Complex Numbers')

# 5. Division
z6 = z1 / z2
print(f'z6 = z1 / z2 = {z6}')
plot_complex_numbers([z1, z2, z6], ['b', 'r', 'g'], ['z1', 'z2', 'z6=z1/z2'], 'Division of Complex Numbers')

# 6. Complex Conjugate
z1bar = np.conj(z1)
print(f'Complex Conjugate of z1 = {z1bar}')

# 7. Opposite of Complex Number
z1opp = -z1
print(f'Opposite of z1 = {z1opp}')

# 8. Polar Coordinates
r1 = abs(z1)  # Magnitude of z1
t1 = np.angle(z1)  # Phase of z1
print(f'Magnitude of z1: r1 = |z1| = {r1}')
print(f'Phase of z1: t1 = angle(z1) = {t1}')

# 9. Polar to Rectangular Conversion
z1_reconstructed = r1 * np.exp(1j * t1)
print(f'z1 from polar form = r1 * exp(j*t1) = {z1_reconstructed}')

# 10. Multiplication Interpretation
z5_interpretation = r1 * abs(z2) * np.exp(1j * (t1 + np.angle(z2)))
print(f'Multiplication Interpretation z5 = z1 * z2 = {z5_interpretation}')

# 11. Division Interpretation
z6_interpretation = r1 / abs(z2) * np.exp(1j * (t1 - np.angle(z2)))
print(f'Division Interpretation z6 = z1 / z2 = {z6_interpretation}')

# 12. Example for complex exponentiation
z7 = 0.707 + 0.707j
z8 = z7**4
z9 = z7**(1/4)
print(f'z7 = {z7}, z7^4 = {z8}, z7^(1/4) = {z9}')

# 13. Mathematical Example for each operation
print("\nMathematical Examples:\n")
# Addition
print("1. Addition Example: z1 + z2 = (3 + 4j) + (1 - 2j) = 4 + 2j")
# Subtraction
print("2. Subtraction Example: z1 - z2 = (3 + 4j) - (1 - 2j) = 2 + 6j")
# Multiplication
print("3. Multiplication Example: z1 * z2 = (3 + 4j) * (1 - 2j) = 3 - 6j + 4j - 8j^2 = 3 - 6j + 4j + 8 = 11 - 2j")
# Division
print("4. Division Example: z1 / z2 = (3 + 4j) / (1 - 2j) = [(3 + 4j)(1 + 2j)] / [(1 - 2j)(1 + 2j)] = (3 + 4j + 6j + 8) / (1^2 + 2^2) = (11 + 10j) / 5 = 2.2 + 2j")
# Conjugate
print("5. Conjugate Example: Conjugate of z1 = 3 - 4j")
# Opposite
print("6. Opposite Example: Opposite of z1 = -3 - 4j")
# Polar to Rectangular
print(f"7. Polar to Rectangular: z1 = {r1} * exp(j * {t1})")

```

## Understanding the 2-Point DFT (Discrete Fourier Transform)

The **Discrete Fourier Transform (DFT)** is a method for converting a sequence of numbers (often time-domain signals) into a different sequence that represents the frequency content of the original sequence. For this example, we will focus on a very simple 2-point DFT.

### 1. What is a 2-Point DFT?

A **2-point DFT** takes a sequence of two numbers and transforms them into their frequency components. In simple terms, you start with two time-domain values (like two measurements taken at different times), and you use the DFT matrix to get the frequency-domain representation.

For the **2-point DFT**, the matrix is:

$$
W_2 = \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix}
$$

This matrix helps calculate the DFT of the input sequence.

### Mathematical Example

Let the input sequence be:

$$
x_n = \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$

This means that the two values in our time-domain signal are both equal to 1.

The DFT transformation can be calculated by multiplying the input sequence by the matrix \( W_2 \). The formula for the DFT is:

$$
X = \frac{1}{N} \cdot W_2 \cdot x_n^T
$$

Where:
- \( N \) is the number of points in the sequence (here, \( N = 2 \)),
- \( x_n^T \) is the transpose of the input sequence, which turns the row vector into a column vector.

Now, let's calculate the DFT:

$$
X = \frac{1}{2} \cdot \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix} \cdot \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$

This gives:

$$
X = \frac{1}{2} \cdot \begin{bmatrix} 2 \\ 0 \end{bmatrix} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}
$$

Thus, the frequency-domain representation of the sequence is:

$$
X = \begin{bmatrix} 1 \\ 0 \end{bmatrix}
$$

This means that the frequency component at the first position is 1, and the frequency component at the second position is 0.

---

### 2. Magnitude and Phase

The **magnitude** is the absolute value of each frequency component, and the **phase** is the angle (or direction) of the component in the complex plane.

For our result:

$$
X = \begin{bmatrix} 1 \\ 0 \end{bmatrix}
$$

- The magnitude of the first component is \( |1| = 1 \),
- The phase of the first component is \( \angle(1) = 0 \).

So, the **magnitude** is 1, and the **phase** is 0.

---

### 3. Parseval’s Identity (Energy in Time and Frequency Domains)

**Parseval’s identity** states that the total energy in the time domain should be equal to the total energy in the frequency domain. It means that the sum of the squares of the values in the time domain should be equal to the sum of the squares of the values in the frequency domain.

#### Energy in the Time Domain:

$$
P_x = \frac{1}{N} \sum_{n=0}^{N-1} |x_n|^2
$$

For our input sequence \( x_n = \begin{bmatrix} 1 \\ 1 \end{bmatrix} \), the energy is:

$$
P_x = \frac{1}{2} \cdot (1^2 + 1^2) = \frac{1}{2} \cdot 2 = 1
$$

#### Energy in the Frequency Domain:

The energy in the frequency domain is the sum of the squares of the frequency components \( X \):

$$
P_{xx} = |X_0|^2 + |X_1|^2 = 1^2 + 0^2 = 1
$$

As you can see, the energy in the time domain (\( P_x = 1 \)) is equal to the energy in the frequency domain (\( P_{xx} = 1 \)), which confirms **Parseval's identity**.

---

### 4. Eigenvalues and Eigenvectors of the DFT Matrix

The **DFT matrix** has special properties that we can explore by calculating its eigenvalues and eigenvectors. These are values that tell us about the matrix's structure.

For our **2-point matrix**:

$$
W_2 = \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix}
$$

#### The Determinant:

The determinant of the matrix tells us if the matrix is invertible (whether it has an inverse).

$$
\text{det}(W_2) = 1 \cdot (-1) - 1 \cdot 1 = -2
$$

This tells us that the matrix is invertible.

#### Eigenvalues and Eigenvectors:

The eigenvalues of \( W_2 \) are \( 2 \) and \( -2 \), and the corresponding eigenvectors are:

- Eigenvectors: \( \begin{bmatrix} 1 \\ 1 \end{bmatrix} \), \( \begin{bmatrix} 1 \\ -1 \end{bmatrix} \)

Eigenvalues and eigenvectors give us insights into how the matrix affects different vectors.

## 5. Matrix Factorizations (LU, QR, SVD)

### Introduction

Matrix factorization methods like **LU**, **QR**, and **Singular Value Decomposition (SVD)** are essential in linear algebra for understanding the structure of matrices and solving linear systems. These factorizations decompose a matrix into simpler, often more useful, matrices.

### 1. LU Factorization

LU factorization decomposes a matrix \( A \) into two matrices:
- **L**: a lower triangular matrix with 1s on the diagonal,
- **U**: an upper triangular matrix.

The LU factorization is written as:
$$
A = L \cdot U
$$

For example, for the matrix \( W_2 \):
$$
W_2 = \begin{bmatrix} 1 & 1 \\ 1 & -1 \end{bmatrix}
$$

The LU factorization results in:
$$
L = \begin{bmatrix} 1 & 0 \\ 1 & 1 \end{bmatrix}, \quad U = \begin{bmatrix} 1 & 1 \\ 0 & -2 \end{bmatrix}
$$

### 2. QR Factorization

QR factorization decomposes a matrix \( A \) into:
- **Q**: an orthogonal matrix (its columns are orthonormal),
- **R**: an upper triangular matrix.

The QR factorization is written as:
$$
A = Q \cdot R
$$

For \( W_2 \), the QR factorization is:
$$
Q = \begin{bmatrix} \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \end{bmatrix}, \quad R = \begin{bmatrix} \sqrt{2} & 0 \\ 0 & -\sqrt{2} \end{bmatrix}
$$

### 3. Singular Value Decomposition (SVD)

SVD decomposes a matrix \( A \) into three matrices:
- **U**: an orthogonal matrix,
- **S**: a diagonal matrix with singular values,
- **V**: an orthogonal matrix.

The SVD factorization is written as:
$$
A = U \cdot S \cdot V^T
$$

For \( W_2 \), the SVD is:
$$
U = \begin{bmatrix} \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \end{bmatrix}, \quad S = \begin{bmatrix} \sqrt{2} & 0 \\ 0 & \sqrt{2} \end{bmatrix}, \quad V = \begin{bmatrix} \frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}} \\ \frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}} \end{bmatrix}
$$


These factorizations are used in various applications like solving linear systems, least squares problems, and dimensionality reduction (e.g., Principal Component Analysis).


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as linalg

# 2-point DFT
N = 2
xn = np.array([1, 1])  # Time sequence

# Fourier 2-point forward transform matrix
W2 = np.array([[1, 1], [1, -1]])

# Forward transform - frequency
X = (np.dot(W2, xn)) / N
magX = np.abs(X)
phaseX = np.angle(X)

# Fourier 2-point inverse transform matrix
W2I = np.conj(W2.T)

# Inverse transform - time
xnn = np.dot(W2I, X)  # Time domain recovery

# Display the results
print("Forward Transform (X):", X)
print("Magnitude of X:", magX)
print("Phase of X:", phaseX)
print("Inverse Transform (xnn):", xnn)

# Plot Magnitude and Phase
n = np.arange(N)
plt.figure()
plt.stem(n, magX, linefmt='r:', markerfmt='r*', basefmt=" ", label='Magnitude')
plt.stem(n, phaseX, linefmt='b-', markerfmt='bo', basefmt=" ", label='Phase')
plt.legend()
plt.grid(True)
plt.title("Magnitude and Phase of 2-point DFT")
plt.show()

# Parseval's Identity: time and frequency domain energy
Px = np.dot(xn, xn) / N
Pxx = np.dot(X, X)

print("\nParseval's Identity (time domain energy):", Px)
print("Parseval's Identity (frequency domain energy):", Pxx)

# Using FFT for comparison
XN = np.fft.fft(xn, N) / N
MAGXN = np.abs(XN)
PHASEXN = np.angle(XN)

print("\nUsing FFT - Magnitude:", MAGXN)
print("Using FFT - Phase:", PHASEXN)

# Determinants, eigenvalues, and eigenvectors
determinantW2 = np.linalg.det(W2)
eigenvaluesW2, eigenvectorsW2 = np.linalg.eig(W2)

print("\nDeterminant of W2:", determinantW2)
print("Eigenvalues of W2:", eigenvaluesW2)
print("Eigenvectors of W2:", eigenvectorsW2)

# LU factorization (requires scipy)
P, LW2, UW2 = linalg.lu(W2)  # Note: P is the permutation matrix
print("\nLU Factorization - P (Permutation matrix):", P)
print("LU Factorization - L:", LW2)
print("LU Factorization - U:", UW2)

# QR factorization
Q, R = np.linalg.qr(W2)
print("\nQR Factorization - Q:", Q)
print("QR Factorization - R:", R)

# Singular Value Decomposition
U, S, V = np.linalg.svd(W2)
print("\nSingular Value Decomposition - U:", U)
print("Singular Value Decomposition - S:", S)
print("Singular Value Decomposition - V:", V)

# Matrix decomposition check
W2W2_reconstructed = np.dot(U, np.dot(np.diag(S), V.T))
print("\nReconstructed W2 from SVD:", W2W2_reconstructed)
```
