# Functions

A function is a relation between a set of inputs and a set of possible outputs, where each input is related to exactly one output.

## Definition of a Function:

A function \( f \) from set \( A \) to set \( B \) is denoted as \( f: A to B \), where each element \( x in A \) is associated with exactly one element \( y in B \), denoted as \( f(x) = y \).

## Examples:

### 1. **Linear function:**

A linear function is of the form:
\[
f(x) = mx + b
\]
where:
- \( m \) is the slope of the line,
- \( b \) is the y-intercept (the value of \( f(x) \) when \( x = 0 \)).

### 2. **Quadratic function:**

A quadratic function is of the form:
\[
f(x) = ax^2 + bx + c
\]
where:
- \( a \), \( b \), and \( c \) are constants,
- The graph of a quadratic function is a parabola.

### 3. **Exponential function:**

An exponential function is of the form:
\[
f(x) = a^x
\]
where:
- \( a \) is a constant (the base of the exponential),
- \( x \) is the exponent.

### 4. **Trigonometric function:**

A trigonometric function is of the form:
\[
f(x) = sin(x)
\]
where:
- \( sin(x) \) is the sine of \( x \),
- \( x \) is typically measured in radians.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f_linear(x):
    return 2*x + 5

def f_quadratic(x):
    return 3*x**2 + 2*x - 1

def f_exponential(x):
    return np.exp(x)

def f_trig(x):
    return np.sin(x)

# Create x values
x = np.linspace(-10, 10, 400)

# Plot the functions
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot Linear function
axs[0, 0].plot(x, f_linear(x))
axs[0, 0].set_title('Linear Function: f(x) = 2x + 5')

# Plot Quadratic function
axs[0, 1].plot(x, f_quadratic(x))
axs[0, 1].set_title('Quadratic Function: f(x) = 3x^2 + 2x - 1')

# Plot Exponential function
axs[1, 0].plot(x, f_exponential(x))
axs[1, 0].set_title('Exponential Function: f(x) = e^x')

# Plot Trigonometric function
axs[1, 1].plot(x, f_trig(x))
axs[1, 1].set_title('Trigonometric Function: f(x) = sin(x)')

plt.tight_layout()
plt.show()
```

## Limits of Functions

**Definition:**

The limit of a function describes the behavior of the function as the input approaches a certain value.

**Limit of a Function at a Point:**

$$ \lim_{x \to c} f(x) = L $$

This means that as $x$ approaches $c$, the function $f(x)$ approaches the value $L$.

**2.1 Limit Laws:**

Some common properties of limits are:

* **Sum Rule:**
    $$ \lim_{x \to c} (f(x) + g(x)) = \lim_{x \to c} f(x) + \lim_{x \to c} g(x) $$
* **Product Rule:**
    $$ \lim_{x \to c} (f(x) \cdot g(x)) = \lim_{x \to c} f(x) \cdot \lim_{x \to c} g(x) $$
* **Quotient Rule:**
    $$ \lim_{x \to c} \frac{f(x)}{g(x)} = \frac{\lim_{x \to c} f(x)}{\lim_{x \to c} g(x)}, \quad \text{if} \quad \lim_{x \to c} g(x) \neq 0 $$

**2.2 Special Limits:**

* **Limit of a Constant:**
    $$ \lim_{x \to c} k = k $$
    The limit of a constant is the constant itself.
* **Limit of $x$ as $x \to c$:**
    $$ \lim_{x \to c} x = c $$
    The limit of $x$ as $x$ approaches $c$ is just $c$.

**2.3 Indeterminate Forms:**

Some limits result in indeterminate forms, such as:

* $\frac{0}{0}$
* $\infty - \infty$
* $0 \times \infty$

These forms may require additional techniques, such as L'Hopital's Rule, to resolve.

**2.4 Example of a Basic Limit:**

Consider the limit:

$$ \lim_{x \to 2} (x^2 - 4) $$

As $x$ approaches 2:

$$ x^2 - 4 = (x - 2)(x + 2) $$

So, we can factor and directly substitute:

$$ \lim_{x \to 2} (x^2 - 4) = (2 - 2)(2 + 2) = 0 $$

```python
import sympy as sp

# Define symbolic variable
x = sp.symbols('x')

# Example 1: Limit of a polynomial function as x approaches 2
f1 = x**2 - 4
limit_f1 = sp.limit(f1, x, 2)
print(f"Limit of f1(x) = x^2 - 4 as x approaches 2: {limit_f1}")

# Example 2: Limit of a trigonometric function as x approaches 0
f2 = sp.sin(x) / x
limit_f2 = sp.limit(f2, x, 0)
print(f"Limit of f2(x) = sin(x) / x as x approaches 0: {limit_f2}")

# Example 3: Limit using L'Hopital's Rule
f3 = (sp.exp(x) - 1) / x
limit_f3 = sp.limit(f3, x, 0)
print(f"Limit of f3(x) = (exp(x) - 1) / x as x approaches 0: {limit_f3}")
```

## Derivatives and Chain Rule

**1. Derivative:**

The derivative of a function measures how the function's output changes as its input changes. It's essentially the rate of change or slope of the function at a specific point.

**Definition of Derivative:**

The derivative of a function $f(x)$ at a point is defined as:

$$ f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h} $$

This is the difference quotient that gives the slope of the tangent line at any point on the function.

**Interpretation:**

* **Rate of change:** How fast a quantity is changing.
* **Slope of the tangent line:** The slope of the line that just touches the curve at a point.

**2. Basic Differentiation Rules:**

**2.1 Power Rule:**

The power rule is used to differentiate functions of the form $f(x) = x^n$, where $n$ is a constant:

$$ \frac{d}{dx}(x^n) = n x^{n-1} $$

**2.2 Sum Rule:**

The sum rule states that the derivative of the sum of two functions is the sum of their derivatives:

$$ \frac{d}{dx}(f(x) + g(x)) = \frac{d}{dx}f(x) + \frac{d}{dx}g(x) $$

**2.3 Product Rule:**

The product rule is used to differentiate the product of two functions:

$$ \frac{d}{dx}(f(x) \cdot g(x)) = f'(x) \cdot g(x) + f(x) \cdot g'(x) $$

**2.4 Quotient Rule:**

The quotient rule is used to differentiate the quotient of two functions:

$$ \frac{d}{dx}\left(\frac{f(x)}{g(x)}\right) = \frac{f'(x) \cdot g(x) - f(x) \cdot g'(x)}{g(x)^2} $$

**2.5 Constant Rule:**

The constant rule states that the derivative of a constant is zero:

$$ \frac{d}{dx}(c) = 0 $$

where $c$ is a constant.

**2.6 Exponential Function Rule:**

The exponential function rule is for differentiating $e^x$:

$$ \frac{d}{dx}(e^x) = e^x $$

**2.7 Trigonometric Functions:**

The derivatives of trigonometric functions are as follows:

* $$ \frac{d}{dx}(\sin(x)) = \cos(x) $$
* $$ \frac{d}{dx}(\cos(x)) = -\sin(x) $$
* $$ \frac{d}{dx}(\tan(x)) = \sec^2(x) $$

**3. Chain Rule:**

The chain rule is a method for differentiating composite functions. If you have a function $y = f(g(x))$, the chain rule tells you how to differentiate this composition.

**Chain Rule Formula:**

If $y = f(g(x))$, then the derivative of $y$ with respect to $x$ is:

$$ \frac{dy}{dx} = f'(g(x)) \cdot g'(x) $$

**Example:**

For $y = \sin(3x)$, applying the chain rule:

* Outer function: $f(u) = \sin(u)$, derivative $f'(u) = \cos(u)$
* Inner function: $g(x) = 3x$, derivative $g'(x) = 3$

The derivative is:

$$ \frac{dy}{dx} = \cos(3x) \cdot 3 = 3 \cos(3x) $$

**4. Higher-Order Derivatives:**

**Second Derivative:**

The second derivative of a function $f(x)$ is the derivative of the derivative:

$$ f''(x) = \frac{d}{dx}(f'(x)) $$

It provides information about the concavity of the function.

**Higher-Order Derivatives:**

These include the third and higher derivatives, useful for motion, optimization, and other applications.

```python
# Python code for Derivatives and Chain Rule

import sympy as sp

# Define the symbol x
x = sp.symbols('x')

# 1. Power Rule: Derivative of x^3
f_power = x**3
f_prime_power = sp.diff(f_power, x)
print(f"Derivative of x^3: {f_prime_power}")

# 2. Sum Rule: Derivative of (x^2 + 3x) + (x^3 - 5x)
f_sum = x**2 + 3*x
g_sum = x**3 - 5*x
sum_derivative = sp.diff(f_sum + g_sum, x)
print(f"Derivative of (x^2 + 3x) + (x^3 - 5x): {sum_derivative}")

# 3. Product Rule: Derivative of (x^2 * sin(x))
f_product = x**2
g_product = sp.sin(x)
product_derivative = sp.diff(f_product * g_product, x)
print(f"Derivative of (x^2 * sin(x)): {product_derivative}")

# 4. Quotient Rule: Derivative of (x^2 / (x + 1))
f_quotient = x**2
g_quotient = x + 1
quotient_derivative = sp.diff(f_quotient / g_quotient, x)
print(f"Derivative of (x^2 / (x + 1)): {quotient_derivative}")

# 5. Constant Rule: Derivative of a constant
c = sp.symbols('c')
constant_derivative = sp.diff(c, x)
print(f"Derivative of constant c: {constant_derivative}")

# 6. Exponential Function Rule: Derivative of e^x
f_exponential = sp.exp(x)
exponential_derivative = sp.diff(f_exponential, x)
print(f"Derivative of e^x: {exponential_derivative}")

# 7. Trigonometric Functions:
# Derivative of sin(x)
f_sin = sp.sin(x)
sin_derivative = sp.diff(f_sin, x)
print(f"Derivative of sin(x): {sin_derivative}")

# Derivative of cos(x)
f_cos = sp.cos(x)
cos_derivative = sp.diff(f_cos, x)
print(f"Derivative of cos(x): {cos_derivative}")

# Derivative of tan(x)
f_tan = sp.tan(x)
tan_derivative = sp.diff(f_tan, x)
print(f"Derivative of tan(x): {tan_derivative}")

# 8. Chain Rule: Derivative of sin(3x)
f_chain = sp.sin(3*x)
chain_derivative = sp.diff(f_chain, x)
print(f"Derivative of sin(3x): {chain_derivative}")
```

## Gradient Descent Overview

Gradient descent is an optimization algorithm used to minimize a function iteratively by adjusting parameters in the direction of the negative gradient. It's widely used in machine learning, particularly for training models.

**Example: Minimizing a Quadratic Function**

Let's use gradient descent to minimize the simple quadratic function:

$$ f(x) = x^2 + 4x + 4 $$

This function represents a simple convex function. The gradient of this function is its derivative.

**Steps in Gradient Descent:**

1.  **Choose a starting point (initial guess $x_0$).**
2.  **Compute the gradient (derivative of the function at the current point).**
3.  **Update the parameters (subtract the learning rate multiplied by the gradient).**
4.  **Repeat until convergence (when the change in the function value is small enough).**

**Function and Gradient:**

Given the function:

$$ f(x) = x^2 + 4x + 4 $$

The gradient (or derivative) is:

$$ f'(x) = 2x + 4 $$

**Gradient Descent Steps:**

1.  **Start at $x = -6$.**
2.  **For each iteration, compute the gradient (derivative), update the value of $x$ by subtracting the learning rate multiplied by the gradient.**
3.  **After 20 iterations, the algorithm converges to the minimum of the function.**

**Output:**

* **Graph:** The plot shows the curve of the function $f(x)$ and the red dots indicate the path taken by the gradient descent algorithm to minimize the function.
* **Final x value:** The printed value of $x$ after 20 iterations will be close to the point where the function reaches its minimum.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function and its derivative
def f(x):
    return x**2 + 4*x + 4

def df(x):
    return 2*x + 4

# Gradient Descent Function
def gradient_descent(learning_rate, initial_x, iterations):
    x_values = [initial_x]  # To track the history of x values
    for i in range(iterations):
        # Compute gradient
        gradient = df(x_values[-1])

        # Update x using the gradient descent rule
        new_x = x_values[-1] - learning_rate * gradient
        x_values.append(new_x)

    return x_values

# Parameters
learning_rate = 0.1  # How big the steps are
initial_x = -6  # Starting point
iterations = 20  # Number of iterations

# Run gradient descent
x_values = gradient_descent(learning_rate, initial_x, iterations)

# Plotting the function and the descent path
x = np.linspace(-10, 2, 400)
y = f(x)

plt.plot(x, y, label='f(x) = x^2 + 4x + 4')
plt.scatter(x_values, f(np.array(x_values)), color='red', zorder=5, label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent on f(x) = x^2 + 4x + 4')
plt.legend()
plt.grid(True)
plt.show()

print(f"Final x value after gradient descent: {x_values[-1]}")
```

## Integrals and Area Under the Curve

An integral is a fundamental concept in calculus that represents the area under a curve. It can be thought of as the reverse process of differentiation. In simple terms, the integral calculates the accumulated area under a function between two points.

**1. Definite Integral:**

The definite integral calculates the area under a curve between two points, say $a$ and $b$:

$$ \int_a^b f(x) \, dx $$

This gives the signed area between the function $f(x)$ and the x-axis from $x = a$ to $x = b$.

**2. Indefinite Integral:**

An indefinite integral is the general form of the integral that doesn't have specific limits. It represents the antiderivative of the function:

$$ \int f(x) \, dx = F(x) + C $$

Where $F(x)$ is the antiderivative and $C$ is the constant of integration.

**Example:**

Consider the function:

$$ f(x) = x^2 $$

**Indefinite Integral of $f(x)$:**

$$ \int x^2 \, dx = \frac{x^3}{3} + C $$

**Definite Integral of $f(x)$ from $x = 1$ to $x = 3$:**

$$ \int_1^3 x^2 \, dx = \left[ \frac{x^3}{3} \right]_1^3 = \frac{3^3}{3} - \frac{1^3}{3} = \frac{27}{3} - \frac{1}{3} = \frac{26}{3} $$

This gives us the area under the curve $f(x) = x^2$ between $x = 1$ and $x = 3$.

#### Explanation of the Code: Integrals and Area Under the Curve

**Indefinite Integral:**

* We calculate the indefinite integral of $f(x) = x^2$ using `sp.integrate(f, x)`.
* The result is $\frac{x^3}{3} + C$, where $C$ is the constant of integration.

**Definite Integral:**

* We calculate the definite integral of $f(x) = x^2$ from $x = 1$ to $x = 3$ using `sp.integrate(f, (x, 1, 3))`.
* The result is $\frac{26}{3}$, which is the area under the curve between these points.

**Plotting:**

* The curve $f(x) = x^2$ is plotted between 0 and 3.
* The area under the curve between $x = 1$ and $x = 3$ is shaded in light blue to visually represent the integral.

**Output:**

The indefinite integral will be printed as:

$$ \int x^2 \, dx = \frac{x^3}{3} + C $$

The definite integral will be printed as:

$$ \int_1^3 x^2 \, dx = \frac{26}{3} $$

The plot will display the function $f(x) = x^2$ with the area under the curve shaded between $x = 1$ and $x = 3$.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the symbol x
x = sp.symbols('x')

# Function f(x) = x^2
f = x**2

# 1. Indefinite Integral
indefinite_integral = sp.integrate(f, x)
print(f"Indefinite Integral of x^2: {indefinite_integral} + C")

# 2. Definite Integral from x = 1 to x = 3
definite_integral = sp.integrate(f, (x, 1, 3))
print(f"Definite Integral of x^2 from 1 to 3: {definite_integral}")

# Plotting the function and the area under the curve
x_vals = np.linspace(0, 3, 400)
y_vals = x_vals**2

# Plotting the curve
plt.plot(x_vals, y_vals, label="f(x) = x^2", color='blue')

# Shading the area under the curve from x=1 to x=3
x_fill = np.linspace(1, 3, 100)
y_fill = x_fill**2
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.4, label='Area under the curve')

# Labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Definite Integral: Area Under the Curve of f(x) = x^2')
plt.legend()
plt.grid(True)
plt.show()
```

## Basic Integral Formulae

Integrals are the foundation of many mathematical concepts and have various rules for evaluating them. Below are some fundamental integral formulae, which are frequently used:

**1. Power Rule:**

For any real number $n \neq -1$, the integral of $x^n$ is:

$$ \int x^n \, dx = \frac{x^{n+1}}{n+1} + C $$

Where $C$ is the constant of integration.

**2. Constant Multiple Rule:**

If $c$ is a constant, the integral of $c \cdot f(x)$ is:

$$ \int c \cdot f(x) \, dx = c \cdot \int f(x) \, dx $$

**3. Sum Rule:**

The integral of a sum of functions is the sum of the integrals:

$$ \int (f(x) + g(x)) \, dx = \int f(x) \, dx + \int g(x) \, dx $$

**4. Exponential Function Rule:**

The integral of $e^x$ is:

$$ \int e^x \, dx = e^x + C $$

**5. Trigonometric Functions:**

The integrals of basic trigonometric functions are:

* $$ \int \sin(x) \, dx = -\cos(x) + C $$
* $$ \int \cos(x) \, dx = \sin(x) + C $$
* $$ \int \sec^2(x) \, dx = \tan(x) + C $$

**6. Basic Integrals Involving Constants:**

* $$ \int 1 \, dx = x + C $$
* $$ \int c \, dx = c \cdot x + C $$ (where $c$ is a constant)

#### Explanation of the Python Code: Basic Integral Formulae

**Power Rule:**

* We compute the integral of $x^n$ where $n = 2$ using `sp.integrate(x**n, x)`.

**Constant Multiple Rule:**

* For the function $3 \cdot x^n$, we compute the integral $\int 3 \cdot x^2 \, dx$ using `sp.integrate(c * x**n, x)`.

**Sum Rule:**

* We compute the integral of $f(x) + g(x)$, where $f(x) = x^2$ and $g(x) = \sin(x)$.

**Exponential Function Rule:**

* The integral of $e^x$ is computed using `sp.integrate(sp.exp(x), x)`.

**Trigonometric Functions:**

* We compute the integrals of $\sin(x)$ and $\cos(x)$.

**Plotting:**

* We visualize the area under the curve for the function $f(x) = x^2$ from $x = 0$ to $x = 3$ using matplotlib.

**Output:**

The indefinite integrals for the above examples will be displayed as follows:

* **Power Rule:** $\int x^2 \, dx = \frac{x^3}{3} + C$
* **Constant Multiple Rule:** $\int 3x^2 \, dx = x^3 + C$
* **Sum Rule:** $\int (x^2 + \sin(x)) \, dx = \frac{x^3}{3} - \cos(x) + C$
* **Exponential Rule:** $\int e^x \, dx = e^x + C$
* **Trigonometric Functions:** $\int \sin(x) \, dx = -\cos(x) + C$, $\int \cos(x) \, dx = \sin(x) + C$

The plot will display the curve of $f(x) = x^2$ along with the shaded area between $x = 0$ and $x = 3$, which represents the area under the curve.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define the symbol x
x = sp.symbols('x')

# 1. Power Rule: x^n dx
n = 2  # Example with n = 2
power_integral = sp.integrate(x**n, x)
print(f"Integral of x^{n}: {power_integral}")

# 2. Constant Multiple Rule: c * f(x) dx
c = 3  # Example with c = 3
constant_multiple_integral = sp.integrate(c * x**n, x)
print(f"Integral of 3 * x^{n}: {constant_multiple_integral}")

# 3. Sum Rule: (f(x) + g(x)) dx
f = x**2
g = sp.sin(x)
sum_integral = sp.integrate(f + g, x)
print(f"Integral of x^2 + sin(x): {sum_integral}")

# 4. Exponential Function Rule: e^x dx
exp_integral = sp.integrate(sp.exp(x), x)
print(f"Integral of e^x: {exp_integral}")

# 5. Trigonometric Functions:
# sin(x) dx
sin_integral = sp.integrate(sp.sin(x), x)
print(f"Integral of sin(x): {sin_integral}")

# cos(x) dx
cos_integral = sp.integrate(sp.cos(x), x)
print(f"Integral of cos(x): {cos_integral}")
```
